```python
#grouping the date to create a dataframe for a induvidual customer statictics
Customer_stats = sales_only.groupby('CustomerID').agg({
    'sales' : 'max',
    'InvoiceDate' : 'max'
})
print(customer_stats.head())
```

```python
#now lwts change the couloum name for the professionality
Customer_stats.columns=['TotalSpend','LastPurchacedDate']
print(customer_stats.head())
```

```python
#lets call teh last date of the purchace as today , so am gonna find the max of the invoice date and call it ass today
today = sales_only['InvoiceDate'].max()
#lets apply logic now sumbtracting last purchaced date by a customer from todays date will give the difference
# #adding a new coloumn called DatesinceLastPurchase
Customer_stats['DateSinceLastPurchase'] = (today - Customer_stats['LastPurchacedDate']).dt.days
print(Customer_stats.head())
```

```python
#finding the ghost vips , ie the customers who spends an lost but hasnt been return in past 60 days
vip_ghost = Customer_stats[(customer_stats['TotalSpend']>1000) & (Customer_stats['DateSinceLastPurchase']>60)]
print(f"Number of the ghost Vip Who Spends a lot of money and hasn,t been retun in 60 days is:{len(vip_ghost)}")
print(f"Total revenue at Risk: {vip_ghost['TotalSpend'].sum():,.2f}")
```

```python
#now we have found we have most higest spenders which is our vip whos hasnt been return since 60 days ir vip ghost
#lets ge their id and find what are tehe top product they purchased 
#so in bussiness perspecticw we could take those customer details and give the like 10% off on the similar or poroduct that they more likely to buy
#lets get the customer ids of this 14 ghost vips
```

```python
ghost_ids = vip_ghost.index
print(ghost_ids)
```

```python
# now lets filter these 14 customers from the main cleaned sales_only data frame
ghost_history = sales_only[sales_only['CustomerID'].isin(ghost_ids)]
print(ghost_history.info())
```

```python
#now lets see the the top 5 products that these ghost vips hasbeen purchased
top_ghost_products = ghost_history['Description'].value_counts().head(5)
print("------TOP PURCHASED PRODUCTS BY GHOST VIPS IS-----")
print(top_ghost_products)
```
